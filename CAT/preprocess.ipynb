```python
# Importing the libraries
import numpy as np
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('AI-DataTrain.csv')
dataset.drop(columns = ['Unnamed: 0'], inplace = True)
dataset.head()
```

```python
# Transposing the dataset
data_trans = dataset.T
data_trans.head()
```

```python
# Calculating the Response Accuracy
sums = data_trans.sum(axis = 1)
response_accuracy = sums/len(list(data_trans.columns))
response_accuracy
```

```python
# Calculating Question Difficulty
diff_level = 1 - response_accuracy

qd = dict(diff_level)
qd_values = list(qd.values())

diff_level
```

```python
# Preparing the Student Ability List
student_ability = []
for i in range(len(dataset)):
    row = dict(dataset.loc[i, :])
    values = list(row.values())
    sa = [qd_values[i] * values[i] for i in range(len(values))]
    cum_sa = np.sum(sa)
    student_ability.append(cum_sa)
```

```python
student_ability
```

```python
# Adding both lists to dataset
data_trans['ResponseAccuracy'] = response_accuracy
data_trans['QuestionDifficulty'] = diff_level
```

```python
# Viewing the updated dataframe
data_trans.head()
```

```python
# Preparing the final dataframe
final_data = data_trans.T
final_data
```

```python
# Adjusting the length of student ability column
student_ability.extend([np.nan,np.nan])

# Adding the student ability column to the final dataframe
final_data['StudentAbility'] = student_ability
```

```python
# Viewing the final dataframe
final_data
```

```python
# Converting to CSV
final_data.to_csv('updated.csv')
```
